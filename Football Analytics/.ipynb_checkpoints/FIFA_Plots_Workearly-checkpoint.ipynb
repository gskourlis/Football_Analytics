```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("/Users/georgiosskourlis/Library/Mobile Documents/com~apple~CloudDocs/Documents/Sports Analytics School/Python/fifa_2023_players.csv")

print (df.head())
```

```python
#Ex1
plt.hist(df['age'],bins=np.arange(df['age'].min(), df['age'].max() + 1, 1), edgecolor='black', alpha=0.7)
plt.show()
```

```python
#Ex2
grouped = df.groupby('league_name').count()['player_id']
grouped.plot(kind="bar", figsize=(15, 5))
plt.show()
```

```python
#Ex3
df.plot.scatter(x='overall',y='value_eur',s=50)
plt.xlabel("Overall")
plt.ylabel("Value in EUR")
plt.title("Relationship between Player Value and Overall Rating")
plt.show()
```

```python
#Ex4
plt.scatter(df['height_cm'], df['weight_kg'],c='r')
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.title("Distribution of Player Heights and Weights")
```

```python
#Ex5
wage_eur = df.groupby("league_name")['wage_eur'].mean()
wage_eur.plot(kind="bar", figsize=(15, 5),color='g')
```

```python
#Ex6
age_group = df.groupby("age")["value_eur"].mean()
plt.plot(age_group, color = "orange")
plt.xlabel("Age")
plt.ylabel("Average Value (EUR)")
plt.title("Average Player Value by Age")
plt.show()
 
```

```python
#Ex7

# Count the number of players in each country
country_counts = df["nationality_name"].value_counts()

# Select the top 10 countries
top_10_countries = country_counts[:10]

 
# Plot the top 10 countries as a bar chart
plt.bar(top_10_countries.index, top_10_countries.values)
 
# Add labels and title to the chart
plt.xlabel("Country")
plt.ylabel("Number of Players")
plt.title("Top 10 Countries with the Most Players in FIFA 2023")
```

```python
#Ex8

skills_rate = df.groupby('skill_moves').size()
plt.pie(skills_rate,autopct='%1.1f%%', startangle=90)
plt.show()
```

```python
#Ex9

# Calculate the average overall score by age
mean_overall_by_age = df.groupby('age')['overall'].mean()
 
# Plot the average overall score by age
plt.plot(mean_overall_by_age.index, mean_overall_by_age, 'o-', color='blue')
 
# Add labels and title to the plot
plt.xlabel('Age')
plt.ylabel('Average Overall Score')
plt.title('Average Player Strength by Age')
 
```

```python
#Ex10

market_value_by_country = df.groupby('nationality_name')['value_eur'].sum().sort_values(ascending=False)
print (market_value_by_country)
```

```python
top_10 = market_value_by_country[:10]

plt.bar(top_10.index,top_10.values)
```
