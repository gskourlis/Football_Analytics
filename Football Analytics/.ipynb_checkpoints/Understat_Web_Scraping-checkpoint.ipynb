### Understat Web Scraping

```python
import re
import json
import requests

from bs4 import BeautifulSoup

import pandas as pd
```

```python
url = 'https://understat.com/match/22169'
```

```python
response = requests.get(url)

print(response.status_code)
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
ugly_soup = str(soup)
```

```python
shots_data = re.search("var shotsData .*= JSON.parse\('(.*)'\)", ugly_soup).group(1)
```

```python
shots_data[:100]
```

```python
data = shots_data.encode('utf8').decode('unicode_escape')
```

```python
data[:100]
```

```python
data = json.loads(data)
```

```python
data
```

```python
home_df = pd.DataFrame(data['h'])
away_df = pd.DataFrame(data['a'])
```
